### Preprocessing

```python
def preprocess_text(text):
    # Keep mentions and hashtags but clean them
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = text.lower()  # Lowercase
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
    return text.strip()
```

```python
spell = SpellChecker()

def selective_correction(word):
    # Correct common misspellings only if they aren't recognized slang
    if spell.unknown([word]):
        corrected_word = spell.correction(word)
        return corrected_word if corrected_word else word
    return word
```

```python
def augment_text(text):
    tokens = text.split()  # TweetTokenizer another option

    # Apply selective correction to common misspellings
    processed_tokens = [selective_correction(word) for word in tokens]
    
    # Introduce minor variations for augmentation
    augmented_texts = []
    for _ in range(3):  # Generate a few augmented versions per text
        augmented_tokens = [
            word if random.random() > 0.2 else selective_correction(word)
            for word in processed_tokens
        ]
        augmented_texts.append(" ".join(augmented_tokens))

    return augmented_texts
```

```python
def preprocess_tweets(df):
    # Apply the preprocess_text function to clean each tweet
    df['TweetText'] = df['TweetText'].apply(preprocess_text)
    
    # Transform text to bag-of-words format for LDA
    tweet_vectors = vectorizer.fit_transform(df['TweetText'])
    
    return tweet_vectors
```

### Loss Function

```python
class MemoryEfficientFocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
    
    def forward(self, inputs, targets):
        ce_loss = torch.nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1-pt)**self.gamma * ce_loss
        return focal_loss.mean()
```

### RoBERTa Model

```python
class GeoTextModel(nn.Module):
    
    def __init__(self, num_regions, char_embedding_dim=64, char_hidden_dim=128, state_feature_dim=16):
        super().__init__()
        
        self.roberta = AutoModel.from_pretrained('roberta-base')
        roberta_hidden_size = 768
        
        self.num_regions = num_regions
        
        # Character-level embedding and processing
        self.char_embedding = nn.Embedding(128, char_embedding_dim)
        self.char_lstm = nn.LSTM(
            char_embedding_dim, char_hidden_dim, num_layers=2, batch_first=True, bidirectional=True, dropout=0.2
        )
        
        self.state_feature_dim = 31
        
        # Attention mechanism with batch_first=True
        self.attention = nn.MultiheadAttention(roberta_hidden_size, num_heads=8, batch_first=True)
        
        # Add state feature dimensions to combined size
        combined_size = roberta_hidden_size + char_hidden_dim * 2 + state_feature_dim #1040
        
        # Classifier with L2 regularization built-in
        self.classifier = nn.Sequential(
            nn.Linear(1055, 512),  # Update input dimension here
            nn.LayerNorm(512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_regions)
        )

        
        # Initialize weights
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.xavier_uniform_(module.weight)
            if module.bias is not None:
                nn.init.zeros_(module.bias)
        elif isinstance(module, nn.LSTM):
            for name, param in module.named_parameters():
                if 'weight' in name:
                    nn.init.xavier_uniform_(param.data)
                elif 'bias' in name:
                    nn.init.zeros_(param.data)

    def forward(self, input_ids, attention_mask, char_ids, state_features):
        # RoBERTa processing
        roberta_outputs = self.roberta(input_ids, attention_mask=attention_mask)
        roberta_sequence = roberta_outputs.last_hidden_state  # Shape: [batch_size, seq_len, hidden_dim]

        # Apply attention mechanism with corrected key_padding_mask
        attn_output, _ = self.attention(
            roberta_sequence, roberta_sequence, roberta_sequence,
            key_padding_mask=~attention_mask.bool()  # Correct shape (batch_size, seq_len)
        )

        # Pool the attention output (mean across sequence dimension)
        roberta_features = torch.mean(attn_output, dim=1)  # Shape: [batch_size, roberta_hidden_size]
        
        # Character-level processing
        char_embeds = self.char_embedding(char_ids)
        char_outputs, (char_hidden, _) = self.char_lstm(char_embeds)
        char_features = torch.cat((char_hidden[-2,:,:], char_hidden[-1,:,:]), dim=1)  # Shape: [batch_size, char_hidden_dim*2]
        
        # Handle state features for validation
        if state_features is None:
            batch_size = input_ids.size(0)
            state_features = torch.zeros(batch_size, self.state_feature_dim, device=input_ids.device)
        
        # Combine all features
        combined_features = torch.cat((roberta_features, char_features, state_features), dim=1)

        # print("roberta_features shape:", roberta_features.shape)
        # print("char_features shape:", char_features.shape)
        # print("state_features shape:", state_features.shape)
        # print("combined_features shape:", combined_features.shape)  # Should be [batch_size, 1055]
        # outputs = self.classifier(combined_features)
        # print("classifier outputs shape:", outputs.shape)  # Should be [32, num_regions]

        # Pass through the classifier
        return self.classifier(combined_features)
```

### Early Stopping

```python
class EarlyStopping:
    def __init__(self, patience=7, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
        
    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0
        
        return self.early_stop
```

### GeoText dataset processing

```python
class GeoTextDataset(Dataset):
    def __init__(self, processed_texts, processed_labels, tokenizer, state_features, max_length=512, max_chars=128):
        self.processed_texts = processed_texts
        self.processed_labels = processed_labels
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.max_chars = max_chars
        self.state_features = state_features
        self.tokenizer = tokenizer
            
    def __len__(self):
        return len(self.processed_texts)
    
    def __getitem__(self, idx):
        text = str(self.processed_texts[idx])
        label = self.processed_labels[idx]
        state_feature = self.state_features[idx]
        
        # Tokenize text
        bert_inputs = self.tokenizer(
            text,
            padding='max_length',
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        # Create character IDs
        char_ids = torch.zeros(self.max_chars, dtype=torch.long)
        for i, c in enumerate(text[:self.max_chars]):
            char_ids[i] = ord(c) % 128
        
        return {
            'input_ids': bert_inputs['input_ids'].squeeze(),
            'attention_mask': bert_inputs['attention_mask'].squeeze(),
            'char_ids': char_ids,
            'labels': torch.tensor(label, dtype=torch.long),
            'state_features': torch.tensor(state_feature, dtype=torch.float32),
        }
```

### Model training parameters

```python
def train_model(model, train_loader, val_loader, device, scaler, num_epochs=10, l2_lambda=0.01):
    criterion = MemoryEfficientFocalLoss(alpha=1, gamma=2)
    optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=l2_lambda)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.1)
    early_stopping = EarlyStopping(patience=5) 

    best_val_loss = float('inf')
    best_val_accuracy = 0.0
    best_model_state = None

    # Create folder to save the model checkpoints
    folder_path = 'model'
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        total_loss = 0.0
        total_correct = 0
        total_samples = 0

        train_pbar = tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs} [Train]', leave=False)
        for batch in train_pbar:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            char_ids = batch['char_ids'].to(device)
            state_features = batch['state_features'].to(device)
            labels = batch['labels'].to(device)

            optimizer.zero_grad()
            
            # Mixed precision training
            with autocast():
                outputs = model(input_ids, attention_mask, char_ids, state_features)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            total_loss += loss.item()
            _, preds = torch.max(outputs, dim=1)
            total_correct += (preds == labels).sum().item()
            total_samples += labels.size(0)

            train_pbar.set_postfix({'loss': f'{loss.item():.4f}'})

        train_accuracy = total_correct / total_samples
        train_loss = total_loss / len(train_loader)

        # Validation phase
        model.eval()
        val_correct = 0
        val_samples = 0
        total_val_loss = 0

        val_pbar = tqdm(val_loader, desc=f'Epoch {epoch + 1}/{num_epochs} [Valid]', leave=False)
        with torch.no_grad():
            for batch in val_pbar:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                char_ids = batch['char_ids'].to(device)
                state_features = batch['state_features'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(input_ids, attention_mask, char_ids, state_features)
                loss = criterion(outputs, labels)
                
                total_val_loss += loss.item()
                _, preds = torch.max(outputs, dim=1)
                val_correct += (preds == labels).sum().item()
                val_samples += labels.size(0)

                val_pbar.set_postfix({'loss': f'{loss.item():.4f}'})

        val_accuracy = val_correct / val_samples
        avg_val_loss = total_val_loss / len(val_loader)

        # Save best model if validation accuracy improves
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            best_model_state = model.state_dict()

        # Scheduler step
        scheduler.step(avg_val_loss)

        # Early stopping check
        if early_stopping(avg_val_loss):
            print("\nEarly stopping triggered")
            break

        print(f"Epoch {epoch + 1}/{num_epochs}, "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.4f}, "
              f"Val Loss: {avg_val_loss:.4f}, Val Acc: {val_accuracy:.4f}")

        # Save model checkpoint at the end of each epoch
        checkpoint_path = os.path.join(folder_path, f"model_checkpoint_epoch_{epoch+1}.pth")
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'train_loss': train_loss,
            'val_loss': avg_val_loss,
            'val_accuracy': val_accuracy
        }, checkpoint_path)
        print(f"Model checkpoint saved at {checkpoint_path}")

    print("Training complete.")
    return best_model_state
```

### Visualization

```python
def create_visualizations(val_labels, val_predictions, state_mapping):
    id_to_state = {v: k for k, v in state_mapping.items()}

    # Confusion Matrix
    plt.figure(figsize=(15, 15))
    conf_matrix = confusion_matrix(val_labels, val_predictions)
    conf_matrix_percent = np.divide(
        conf_matrix, conf_matrix.sum(axis=1, keepdims=True), 
        out=np.zeros_like(conf_matrix, dtype=float), where=conf_matrix.sum(axis=1, keepdims=True) != 0
    ) * 100
    sns.heatmap(conf_matrix_percent,
                xticklabels=[id_to_state[i] for i in range(len(state_mapping))],
                yticklabels=[id_to_state[i] for i in range(len(state_mapping))],
                annot=True, fmt='.1f', cmap='Blues')
    plt.title('State Prediction Confusion Matrix (%)')
    plt.xlabel('Predicted State')
    plt.ylabel('True State')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Per-State Accuracy
    state_metrics = {}
    for state_id in range(len(state_mapping)):
        state_name = id_to_state[state_id]
        mask = (val_labels == state_id)
        if mask.sum() > 0:
            accuracy = (val_predictions[mask] == state_id).mean()
            total_predictions = (val_predictions == state_id).sum()
            total_actual = mask.sum()
            state_metrics[state_name] = {
                'Accuracy': accuracy,
                'Total Predictions': total_predictions,
                'Total Actual': total_actual
            }

    metrics_df = pd.DataFrame(state_metrics).T
    plt.figure(figsize=(15, 6))
    ax = metrics_df['Accuracy'].plot(kind='bar')
    plt.title('Prediction Accuracy by State')
    plt.xlabel('State')
    plt.ylabel('Accuracy')
    plt.axhline(y=metrics_df['Accuracy'].mean(), color='r', linestyle='--', 
                label=f'Mean Accuracy: {metrics_df["Accuracy"].mean():.3f}')
    plt.legend()
    for i, v in enumerate(metrics_df['Accuracy']):
        ax.text(i, v, f'{v:.3f}', ha='center', va='bottom')
    plt.tight_layout()
    plt.show()

    # Number of Samples per State
    plt.figure(figsize=(15, 6))
    metrics_df['Total Actual'].plot(kind='bar')
    plt.title('Number of Samples per State')
    plt.xlabel('State')
    plt.ylabel('Number of Samples')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

val_labels = df['state_id'].values
val_predictions = np.random.choice(val_labels, size=len(val_labels))
create_visualizations(val_labels, val_predictions, state_mapping)
```
